### Calculating Interest Earned
>let's calculate interest payments and FV with python script.

Suppose that two years after you place 1000 USD in a deposit accouoont, the value of the deposit grows to 1200 USD. Build python script that calculates teh annual interest rate. 

$FV = PV(1+r)^t$

$r = (\frac{FV}{PV})^{1/t}-1$

>You can use python like a calculator and conveniently, you can define variables before using them in an equation. This is the strategy that we'll take for solving the above equation

```python
FV = 1200
PV = 1000
t = 2
r = (FV/PV)**(1/t)-1
r
```

>Let's test this result to make sure that $PV(1+r)^t$ yields the correct FV.

```python
FV_test = PV*(1+r)**t
round(FV_test,2) #rounds the result to 2 decimal places
```

Interest can be compounded annually (once per year), or more than once. If the interest rate is compounded more than once annually, in the PV equation, we divide the interest by the number of times compounded, $n$ and we multiply the number of years, $t$, by $n$. 

>$FV = PV(1+\frac{1}{n})^{tn}$ 

Next, create a dictionary that records the value of $1000 after a year of collecting interest over different rates of compounding. 

```python
#use dictionary to define how many times interest is compounded
compounding_def = {"annual":1,
                  "semi-annual":2,
                  "quarterly": 4,
                  "monthly": 12,
                  "weekly": 52,
                  "daily": 365}
#compound interest at rate of 10% which is same as 0.1
PV = 1000
r = 0.1
t = 1

#create dictionary to hold the results
compounding_interest = {}

#for var1. var2. om dict.items()
#for key1, value1 in dict.items() so this calls the key and the item it's pointing to
for period_name, n in compounding_def.items():
    
    #use round(val, #decimals) to round value to a specified # of decimals
    compounding_interest[period_name] = round(PV * (1 + r/n) ** (t*n), 2)
compounding_interest
```

Now let's use a for loop to calculate the FV of deposits for interest rates from 1% to 20%

```python
compounding_def = {"annual":1,
                  "semi-annual":2,
                  "quarterly": 4,
                  "monthly": 12,
                  "weekly": 52,
                   "daily": 365}
#create/define variables variables
PV = 1000
t = 2

#use generator function to create a list of interest rates from 1 to 20
rates = [i for i in range(1,21)] #this says use i to enter values in the list
                                    #also interprete this as, you have some input i, then create entries from i=1 to i=20
compounding_interest = {}
for rate in rates:
    r = rate/100
    #let's make the rate a string
    rate = str(rate)+"%"
    compounding_interest[rate] = {} #for every interest rate, we'll create an entry that is a dictionary
    for period_name, n in compounding_def.items():
        compounding_interest[rate][period_name] = round(PV * (1 + r / n) ** (t*n),2)
compounding_interest
```

```python
compounding_interest_df = pd.DataFrame(compounding_interest)
compounding_interest_df
```

>Let's look at another example. Our goal is to teach the program to search for a solution to the following algebra problem.

```python
a = 3
x = 1
ax = a * x
#solve for ax = 1000
margin = 1.0 #this tells the program to increase/decrease the solution by a float, rather than an integer. 

#create a variable that detects if we are above or below the desired answer, which in this case is 1000
below_1000 = True          #if we go from below to 1000, decrease the size of the change and
decrease_margin = False    #switch the direction of the change, so change from positive to negative

while round(ax, 10) != 1000:    
    if decrease_margin:
        margin = margin * 0.1
        decrease_margin = False
   
    # increase x or decrease x depending on whether
    #x is below or above 1000
    x = x + margin if ax < 1000 else x - margin
    ax = a * x
    print("x =", x, "\tax", ax)        #\t means tab so \tax means tab then ax
   
    # if ax crosses over 1000, identify
    # 1. taht ax is now greater than 1000
    # 2. x should change by a smaller margin and in a different direction
    if below_1000 == True and ax > 1000:
        #change by smaller margin
        decrease_margin = True
        #move in opposite direction
        below_1000 = False
    elif below_1000 == False and ax<1000:
        #change by smaller margin
        decrease_margin== True
        #move in opposite direction
        below_1000 = True
print("Solution: (" + str(a) + "("+ str(x) +") =", ax)
    
```

```python
#solve ax = 1000
a = 3

# we don't know x, so we will count x upward beginning from 1
x = 1
ax = a * x

#we know that we want to count x upward until ax > 1000,
#then we want to count downward at a lower rate
#i.e, first count upward by 1, then count downward by 0.1,
# then upward again by 0.001, etc.

margin = 1.0

#create variables to keep track of ax's value relative to the target solution of 1000
below_1000_currently = True
#we also need to know whether ax was above or below 1000 last period, to know whether or not the direction
#of the change in x needs to switch
below_1000_last_period = True

#if ax is approx. 1000, break out of while loop
while round(ax, 10) != 1000:
    #if ax<1000, count upward
    #if ax>1000, count downward
    
    x = x + margin if ax<1000 else x - margin
    #after changing the value x, update the value of ax
    ax = a * x
    #update whether ax is above or below 1000
    below_1000_currently = True if ax<1000 else False
    #if ax crossed from above 1000 to below 1000
    #or if ax crossed from below 1000 to above 1000, 
    #shrink the margin by which we change x
    if below_1000_currently == True and below_1000_last_period == False:
        margin = margin * 0.1
        print("start counting upward at range of", margin)
        
    if below_1000_currently == False and below_1000_last_period == True:
        margin = margin * 0.1
        print("Start counting downward at rate of", margin)
    
    #once we have identified if ax has crossed 1000, 
    #update last period's position to match the current
    below_1000_last_period = below_1000_currently
    print("x:", x, "ax:", ax)
    
print("You solved ax = 1000, a = 3, for x")
print("x=", round(x, 2), "ax =", round(ax, 2))
```

## Using brute force to solve for the interest rate:
>In this problem, we would need to use code to grow 1000 USD to 1500 USD after 2 years of annual compounding

```python
# define variables that will be used in the equation
# equation for PV = FV/(1+r/n)**(t*n)
# equation for FV = PV(1+r/n)**(t/n)

PV = 1000
FV = 1000
# checking all interest rates from 0.01 (1%) and above
r = 0.01

# margin should increase/decrease by no more than 1%
margin = 0.1

# Identify whether FV is above or below 1500
below_1500 = True

# set decrease_margin to False
# decrease_margin will only be True if either 
#                      below_1500 == True and FV > 1500
#                      or
#                      below_1500 == False and FV < 1500
# once decrease_margin is true, margin shrinks and decrease_margin is again set to False
decrease_margin = False

while round(FV, 2) != 1500.00:
    if decrease_margin:          #means if decrease_margin is true
        margin = margin * 0.1
        decrease_margin = False
    # r increases by margin if below 1500
    # decreases by margin if above 1500
    r = r + margin if FV < 1500 else r - margin
    
    #recalculate FV given change in r
    FV = PV * (1 + r)**2
    print("r:", r, "\tFV:", FV)
    if below_1500 == True and FV > 1500:
        decrease_margin = True
        below_1500 = False
    if below_1500 == False and FV < 1500:
        decrease_margin = True
        below_1500 = True
print("r:", r, "FV:", round(FV, 2))
```

```python
# define variables that will be used in the equation
# equation for PV = FV/(1+r/n)**(t*n)
# equation for FV = PV(1+r/n)**(t/n)

PV = 1000
FV = 1000
# checking all interest rates from 0.01 (1%) and above
r = 0.01

# margin should increase/decrease by no more than 1%
margin = 0.1

# Identify whether FV is above or below 1500
below_1500_current = True
below_1500_last_period = True


# decrease margin if either: 
#                      below_1500_last_period == True and below_1500_current == False
#                      or
#                      below_1500_last_period == False and below_1500_current == True

while round(FV, 2) != 1500.00:
    # r increases by margin if below 1500
    # r decreases by margin if above 1500
    r = r + margin if FV < 1500 else r - margin
    
    #recalculate FV given change in r
    FV = PV * (1 + r)**2
    below_1500_current = True if FV < 1500 else False
    print("r:", r, "\tFV:", FV)
    
    if below_1500_last_period == True and below_1500_current == False:
        margin = margin * 0.1
        below_1500 = False
    if below_1500_last_period == False and below_1500_current == True:
        margin = margin * 0.1
    below_1500_last_period = below_1500_current
    
print("r:", r, "FV:", round(FV, 2))
```

### Now use a dictionary and brute force to solve for the interest rate you would need to grow 1000 to each value from 1100 to 2000 after 2 years of annual compounding

```python
PV = 1000
FV_dict = {}       #The creates a dictionary that can be used to save the solution later on

for final_FV in range(1100, 2001, 100):
    r  = 0.01
    margin = 0.01  
    below_final_FV = True
    decrease_margin = False
    
    while round(FV, 2) != final_FV:
        if decrease_margin:
            margin = margin * 0.1
            decrease_margin = False
        r = r + margin if FV < final_FV else r - margin
        FV =  PV * (1 + r) **2
        if below_final_FV == True and FV > final_FV:
            decrease_margin = True
            below_final_FV = False
        if below_final_FV == False and FV < final_FV:
            decrease_margin = True
            below_final_FV = True
    FV_dict[final_FV] = r                           #this saves the values in the dictionary
FV_dict
```
